# Cleaning Data

kesalahan tipe data pada order_df

```python
datetime_columns = ["order_purchase_timestamp",
                    "order_approved_at",
                    "order_delivered_carrier_date",
                    "order_delivered_customer_date",
                    "order_estimated_delivery_date"]

for column in datetime_columns:
  order_df[column] = pd.to_datetime(order_df[column])
```

```python
order_df.info()
```

missing value pada order_df

```python
order_df = order_df.dropna()
print(order_df.isnull().sum())
```

missing value pada product_df

```python
product_df = product_df.dropna()
print(product_df.isnull().sum())
```

Kesalahan tipe data pada orderview_df

```python
datetime_columns = ["review_creation_date",
                    "review_answer_timestamp"]

for column in datetime_columns:
  orderview_df[column] = pd.to_datetime(orderview_df[column])
```

```python
orderview_df.info()
```

missing value pada orderview_df

```python
orderview_df = orderview_df.dropna()
print(orderview_df.isnull().sum())
```

# **Data Visualization & Explanatory**

Diantara semua metode pembayaran, metode apa yang paling sering digunakan oleh customer?

```python
payment =orderpay_df.groupby(by="payment_type").order_id.nunique().sort_values(ascending=False)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
payment.plot(kind="bar", color = "purple")
plt.title('order dengan metode pembayaran terbanyak')
plt.xlabel('metode pembayaran')
plt.ylabel('jumlah order')
plt.show()
```

Bagaimana performa penjualan di tiap bulannya?

```python
product =product_df.groupby(by="product_category_name").product_id.nunique().sort_values(ascending=False)
monthly_df = order_df.resample(rule='M', on='order_purchase_timestamp').agg({
    "order_id": "nunique"
})
monthly_df.index = monthly_df.index.strftime('%B') #mengubah format order date menjadi Tahun-Bulan
monthly_df = monthly_df.reset_index()
monthly_df.head()

plt.figure(figsize=(10,5))
sns.histplot(monthly_df['order_purchase_timestamp'], kde=True, color='purple')
plt.title('Performa Penjualan selama setahun')
plt.xticks(rotation=45)
plt.xlabel('Bulan')
plt.ylabel('Frekuensi')
plt.show()
```
